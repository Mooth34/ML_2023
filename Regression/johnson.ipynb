```python
import random

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('datasets/JohnsonJohnson.csv')
y = data['value']
X = data['time']
# data split
X_1 = (np.arange(int(len(X) / 4)) + 1960).reshape(int(len(y) / 4), 1)
X_2 = (np.arange(int(len(X) / 4)) + 1960.25).reshape(int(len(y) / 4), 1)
X_3 = (np.arange(int(len(X) / 4)) + 1960.5).reshape(int(len(y) / 4), 1)
X_4 = (np.arange(int(len(X) / 4)) + 1960.75).reshape(int(len(y) / 4), 1)
X_year = (np.arange(int(len(X) / 4)) + 1960).reshape(int(len(y) / 4), 1)
y_1 = y[(data['time'] % 1 == 0)]
y_2 = y[(data['time'] % 1 == 0.25)]
y_3 = y[(data['time'] % 1 == 0.5)]
y_4 = y[(data['time'] % 1 == 0.75)]
y_year = []
for i1, i2, i3, i4 in zip(y_1, y_2, y_3, y_4):
    y_year.append((i1 + i2 + i3 + i4) / 4)
```

```python
# 1 Quarter
X_1train, X_1test, y_1train, y_1test = train_test_split(X_1, y_1, random_state=random.randint(0, 1000))
lmr = LinearRegression().fit(X_1train, y_1train)
y_pred = lmr.predict(X_1test)
print('linear 1 train {}'.format(lmr.score(X_1train, y_1train)))
print('linear 1 test {}'.format(lmr.score(X_1test, y_1test)))
print('linear 1 mse test {}'.format(mean_squared_error(y_1test, y_pred)))
```

```python
q1_pred = lmr.predict(X_1)
plt.xlabel('Dates')
plt.ylabel('Income')
plt.plot(X_1, q1_pred, label='DAX', marker='.', markersize=1)
plt.plot(X_1, y_1, label='SMI', marker='.', markersize=1)
plt.title(' 1st Quarter')
plt.grid()
plt.show()
```

```python
print('predict 1q 2016 : {}'.format(lmr.predict([[2016]])))
```

```python
# year
X_ytrain, X_ytest, y_ytrain, y_ytest = train_test_split(X_year, y_year, random_state=random.randint(0, 1000))
lmr = LinearRegression().fit(X_ytrain, y_ytrain)
y_pred = lmr.predict(X_ytest)
print('linear year train {}'.format(lmr.score(X_ytrain, y_ytrain)))
print('linear year test {}'.format(lmr.score(X_ytest, y_ytest)))
print('linear year mse test {}'.format(mean_squared_error(y_ytest, y_pred)))
```

```python
year_pred = lmr.predict(X_year)
plt.xlabel('Dates')
plt.ylabel('Volumes')
plt.plot(X_year, year_pred, label='DAX', marker='.', markersize=1)
plt.plot(X_year, y_year, label='SMI', marker='.', markersize=1)
plt.title(' Year')
plt.grid()
plt.show()
```
