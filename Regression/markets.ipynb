```python
import random

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('datasets/EuStockMarkets.csv')
y = data.loc[:, 'DAX':'FTSE']
X = np.arange(len(y['DAX'])).reshape(len(y['DAX']), 1)
# display all dependencies
plt.xlabel('Days')
plt.ylabel('Values')
plt.plot(y['DAX'], label='DAX', marker='.', markersize=1)
plt.plot(y['SMI'], label='SMI', marker='.', markersize=1)
plt.plot(y['CAC'], label='CAC', marker='.', markersize=1)
plt.plot(y['FTSE'], label='FTSE', marker='.', markersize=1)
plt.legend()
plt.grid()
plt.show()
```

```python
print(X)
print(y)
```

```python
y_train, y_test, X_train, X_test = train_test_split(y, X, random_state=random.randint(0, 1000))

# least squares Linear Regression DAX
lmr = LinearRegression().fit(X_train, y_train['DAX'])
y_pred = lmr.predict(X_test)
print('linear train {}'.format(lmr.score(X_train, y_train['DAX'])))
print('linear DAX test {}'.format(lmr.score(X_test, y_test['DAX'])))
print('linear mse test {}'.format(mean_squared_error(y_test['DAX'], y_pred)))
```

```python
DAX = lmr.predict(X)
plt.xlabel('Days')
plt.ylabel('Values')
plt.plot(y['DAX'], label='DAX', marker='.', markersize=1)
plt.plot(DAX, label='DAX', marker='.', markersize=1)
plt.title('DAX')
plt.grid()
plt.show()
```

```python
# least squares Linear Regression SMI
lmr = LinearRegression().fit(X_train, y_train['SMI'])
y_pred = lmr.predict(X_test)
print('linear SMI train {}'.format(lmr.score(X_train, y_train['SMI'])))
print('linear SMI test {}'.format(lmr.score(X_test, y_test['SMI'])))
print('linear SMI mse test {}'.format(mean_squared_error(y_test['SMI'], y_pred)))
```

```python
SMI = lmr.predict(X)
plt.xlabel('Days')
plt.ylabel('Values')
plt.plot(y['SMI'], label='DAX', marker='.', markersize=1)
plt.plot(SMI, label='DAX', marker='.', markersize=1)
plt.title('SMI')
plt.grid()
plt.show()
```

```python
# least squares Linear Regression CAC
lmr = LinearRegression().fit(X_train, y_train['CAC'])
y_pred = lmr.predict(X_test)
print('linear CAC train {}'.format(lmr.score(X_train, y_train['CAC'])))
print('linear CAC test {}'.format(lmr.score(X_test, y_test['CAC'])))
print('linear CAC mse test {}'.format(mean_squared_error(y_test['CAC'], y_pred)))
```

```python
CAC = lmr.predict(X)
plt.xlabel('Days')
plt.ylabel('Values')
plt.plot(y['CAC'], label='DAX', marker='.', markersize=1)
plt.plot(CAC, label='DAX', marker='.', markersize=1)
plt.title('CAC')
plt.grid()
plt.show()
```

```python
# least squares Linear Regression FTSE
lmr = LinearRegression().fit(X_train, y_train['FTSE'])
y_pred = lmr.predict(X_test)
print('linear FTSE train {}'.format(lmr.score(X_train, y_train['FTSE'])))
print('linear FTSE test {}'.format(lmr.score(X_test, y_test['FTSE'])))
print('linear FTSE mse test {}'.format(mean_squared_error(y_test['FTSE'], y_pred)))
```

```python
FTSE = lmr.predict(X)
plt.xlabel('Days')
plt.ylabel('Values')
plt.plot(y['FTSE'], label='DAX', marker='.', markersize=1)
plt.plot(FTSE, label='DAX', marker='.', markersize=1)
plt.title('FTSE')
plt.grid()
plt.show()
```
