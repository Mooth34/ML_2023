```python
import random

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('datasets/Sunspots.csv')
y = np.array(data['Monthly Mean Total Sunspot Number']).reshape(len(data['Monthly Mean Total Sunspot Number']), 1)
X = np.array(data['Unnamed: 0']).reshape(len(y), 1)

X_1train, X_1test, y_1train, y_1test = train_test_split(X, y, random_state=random.randint(0, 1000))
lmr = LinearRegression().fit(X_1train, y_1train)
y_pred = lmr.predict(X_1test)
print('linear train {}'.format(lmr.score(X_1train, y_1train)))
print('linear test {}'.format(lmr.score(X_1test, y_1test)))
print('linear mse test {}'.format(mean_squared_error(y_1test, y_pred)))
```

```python
y_pred = lmr.predict(X)
plt.xlabel('Dates')
plt.ylabel('Monthly Mean Total Sunspot Number')
plt.plot(X, y_pred, label='DAX', marker='.', markersize=1)
plt.plot(data['Date'], y, label='SMI', marker='.', markersize=1)
plt.grid()
plt.show()
```
