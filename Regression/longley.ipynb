```python
import random

import pandas as pd
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

from display_researching import display_ridge_alpha_train_test
```

```python
data = pd.read_csv('datasets/longley.csv')
X = data.loc[:, 'GNP.deflator':'Year']
y = data['Employed']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=random.randint(0, 1000))

# least squares Linear Regression with Population
lmr = LinearRegression().fit(X_train, y_train)
y_pred = lmr.predict(X_test)
print('linear with Population train {}'.format(lmr.score(X_train, y_train)))
print('linear with Population test {}'.format(lmr.score(X_test, y_test)))
print('linear with Population mse test {}'.format(mean_squared_error(y_test, y_pred)))
```

```python
# ridge Linear Regression with Population
display_ridge_alpha_train_test(X, y)
```

```python
rlr = Ridge(alpha=0.1).fit(X_train, y_train)
y_pred = rlr.predict(X_test)
print('ridge with Population train {}'.format(rlr.score(X_train, y_train)))
print('ridge with Population test {}'.format(rlr.score(X_test, y_test)))
print('ridge with Population mse test {}'.format(mean_squared_error(y_test, y_pred)))
```

```python
del X['Population']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=random.randint(0, 1000))
# ridge Linear Regression without Population
display_ridge_alpha_train_test(X, y)
```

```python
rlr = Ridge(alpha=0.1).fit(X_train, y_train)
y_pred = rlr.predict(X_test)
print('ridge without Population train {}'.format(rlr.score(X_train, y_train)))
print('ridge without Population test {}'.format(rlr.score(X_test, y_test)))
print('ridge without Population mse test {}'.format(mean_squared_error(y_test, y_pred)))
```
