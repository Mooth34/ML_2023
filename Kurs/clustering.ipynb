```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.cluster.hierarchy as sch
from sklearn.cluster import KMeans
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('datasets/heart.csv')
# X = data.loc[:, 'age':'thal']

X = data.loc[:, ['cp', 'slope', 'restecg', 'thalach']]
y = data['target']
k_means = KMeans(n_clusters=2, init='random')
k_means.fit(X)

def sum_err(labels, y):
    sum = 0
    for i in range(len(labels)):
        if labels[i] == y[i]:
            sum += 1
    return sum / len(labels)
```

```python
print('mse clustering{}'.format(mean_squared_error(y, k_means.labels_)))
print('right clustering {}'.format(sum_err(k_means.labels_, y)))
```

```python
plt.figure()
dendogram = sch.dendrogram(sch.linkage(data.loc[:, 'age':'thal']), p=8, truncate_mode='level')
print(dendogram.values())
plt.grid()
plt.show()
```
