```python
import random

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier

from neural_researching import display_dependency_for_layers, \
display_dependency_for_function, display_dependency_for_iter

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('datasets/heart.csv')
y = data['target']
X = data.loc[:, 'age':'thal']
```

```python
display_dependency_for_layers(X, y, 50)
```

```python
display_dependency_for_layers(X, y, 100)
```

```python
display_dependency_for_layers(X, y, 150)
```

```python
display_dependency_for_function(X, y, 150)
```

```python
display_dependency_for_iter(X, y)
```

```python
mean_accuracy_test = 0.
mean_accuracy_train = 0.
test_count = 30
for _ in np.arange(test_count):
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=random.randint(0, 1000))
    clf = MLPClassifier(solver='lbfgs', activation='relu', hidden_layer_sizes=(50, 50), max_iter=150)
    clf.fit(X_train, y_train)
    mean_accuracy_test += clf.score(X_test, y_test) / test_count
    mean_accuracy_train += clf.score(X_train, y_train) / test_count

print("train score {}".format(mean_accuracy_train))
print("test score {}".format(mean_accuracy_test))
```
