```python
import random

import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

from accuracy_dependence import get_accuracy_dependence_on_the_k, get_accuracy_dependence_on_the_metric, \
display_overtraining_for_k, display_overtraining_for_p
```

```python
data = pd.read_csv('datasets/heart.csv')
y = data['target']
X = data.loc[:, 'age':'thal']
```

```python
accuracy_dependence_on_the_uniform = get_accuracy_dependence_on_the_k(X=X, y=y, uniform='uniform')
accuracy_dependence_on_the_distance = get_accuracy_dependence_on_the_k(X=X, y=y, uniform='distance')
plt.plot(accuracy_dependence_on_the_uniform, label='Uniform', marker='.', markersize=10)
plt.plot(accuracy_dependence_on_the_distance, label='Distance', marker='.', markersize=10)
plt.xlabel('K')
plt.ylabel('Misclass')
plt.legend()
plt.grid()
plt.show()
```

```python
accuracy_dependence_on_the_uniform = get_accuracy_dependence_on_the_metric(X=X, y=y, uniform='uniform')
accuracy_dependence_on_the_distance = get_accuracy_dependence_on_the_metric(X=X, y=y, uniform='distance')

plt.plot(accuracy_dependence_on_the_uniform, label='Uniform', marker='.', markersize=10)
plt.plot(accuracy_dependence_on_the_distance, label='Distance', marker='.', markersize=10)
plt.xlabel('P')
plt.ylabel('Misclass')
plt.legend()
plt.grid()
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=random.randint(0, 1000))
clf = KNeighborsClassifier(n_neighbors=23, weights='uniform', p=1)
clf.fit(X_train, y_train)
print('train score: {}'.format(clf.score(X_train, y_train)))
print('test score: {}'.format(clf.score(X_test, y_test)))
```
