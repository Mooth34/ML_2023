# Задание 1.1. Крестики-нолики

```python
import matplotlib.pyplot as plt
import pandas as pd
from accuracy_dependence import get_accuracy_dependence_on_the_sample_size, \
        get_accuracy_dependence_on_the_test_size, normalize_data
```

## Чтение данных из файла

```python
data = pd.read_csv('datasets/Tic_tac_toe.txt', header=None)
y = data[9]
X = data.loc[:, '0':'8'].isin([1.0, 'x'])
X = normalize_data(X)
```

```python
# Зависимость точности от объема выборки
accuracy_dependence_on_the_sample_size = get_accuracy_dependence_on_the_sample_size(X=X, y=y)
```

```python
# Зависимость точности от объема тестовых данных
accuracy_dependence_on_the_test_size = get_accuracy_dependence_on_the_test_size(X=X, y=y,
        train_size=int(y.index.size * 0.6))
```

## Графическое отображение результата

```python
plt.plot(accuracy_dependence_on_the_test_size, label='Test', marker='.', markersize=10)
plt.xlabel('Test size')
plt.grid()
plt.show()
```

```python
plt.plot(accuracy_dependence_on_the_sample_size, label='Sample', marker='.', markersize=10)
plt.xlabel('Sample size')
plt.ylabel('Accuracy')
plt.grid()
plt.show()
```
