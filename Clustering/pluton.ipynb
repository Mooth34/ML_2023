```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('datasets/pluton.csv')

for iter in [1, 2, 10]:
    k_means = KMeans(n_clusters=3, max_iter=iter)
    k_means.fit(data)
```

```python
plt.scatter(data['Pu238'], data['Pu239'], c=k_means.labels_, alpha=0.4)
plt.scatter(k_means.cluster_centers_[:, 0], k_means.cluster_centers_[:, 1], c='red', alpha=0.7)
plt.xlabel('Pu238')
plt.ylabel('Pu239')
plt.title('iter count = {}'.format(k_means.n_iter_))
plt.grid()
plt.show()
```

```python
plt.scatter(data['Pu240'], data['Pu241'], c=k_means.labels_, alpha=0.4)
plt.scatter(k_means.cluster_centers_[:, 2], k_means.cluster_centers_[:, 3], c='red', alpha=0.7)
plt.xlabel('Pu240')
plt.ylabel('Pu241')
plt.title('iter count = {}'.format(k_means.n_iter_))
plt.grid()
plt.show()
```
