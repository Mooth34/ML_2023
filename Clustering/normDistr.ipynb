```python
import matplotlib.pyplot as plt
from numpy.random._generator import default_rng
from sklearn_extra.cluster import KMedoids
import numpy as np
```

```python
data = np.zeros((300, 3))
rng = default_rng()
for i in range(100):
    data[3*i] = [rng.standard_normal()*100, rng.standard_normal(), rng.standard_normal()]
    data[3*i+1] = [rng.standard_normal(), rng.standard_normal()*30, rng.standard_normal()]
    data[3*i+2] = [rng.standard_normal(), rng.standard_normal(), rng.standard_normal()*70]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data[:, 0], data[:, 1], data[:, 2])
plt.grid()
plt.show()
```

```python
# for method in ['alternate', 'pam']:
for metric in ['euclidean', 'cosine', 'cityblock', 'manhattan', 'l1', 'l2']:
    k_medoids = KMedoids(n_clusters=3, init='k-medoids++', metric=metric)
    k_medoids.fit(data)
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(data[:, 0], data[:, 1], data[:, 2], c=k_medoids.labels_, alpha=0.5)
    ax.scatter(k_medoids.cluster_centers_[:, 0], k_medoids.cluster_centers_[:, 1], k_medoids.cluster_centers_[:, 2],
    c='red', alpha=0.7)
    plt.title('metric = {}'.format(metric))
    plt.grid()
    plt.show()
```
