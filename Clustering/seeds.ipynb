```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.cluster.hierarchy as sch
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('datasets/seeds_dataset.txt', sep='\s+', header=None)

for iter in [1, 10, 40]:
    k_means = KMeans(n_clusters=3, max_iter=iter)
    k_means.fit(data)

plt.scatter(data[0], data[1], c=k_means.labels_, alpha=0.5)
plt.xlabel('0')
plt.ylabel('1')
plt.title('iter count = {}'.format(k_means.n_iter_))
plt.grid()
plt.show()
```

```python
plt.scatter(data[2], data[3], c=k_means.labels_, alpha=0.5)
plt.xlabel('2')
plt.ylabel('3')
plt.grid()
plt.show()
```

```python
plt.scatter(data[4], data[5], c=k_means.labels_, alpha=0.5)
plt.xlabel('4')
plt.ylabel('5')
plt.grid()
plt.show()
```

```python
plt.figure()
animal_dendogram = sch.dendrogram(sch.linkage(data.loc[:, [0, 1, 2, 3, 4, 5, 6]]), p=8, truncate_mode='level')
print(animal_dendogram.values())
plt.grid()
plt.show()
```
