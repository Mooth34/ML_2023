```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.cluster.hierarchy as sch
from sklearn.cluster import KMeans  #AgglomerativeClustering
```

```python
# task 3
data = pd.read_csv('datasets/votes.repub.csv')
data.rename(columns={'Unnamed: 0': 'City'}, inplace=True)
data.fillna(method='backfill', inplace=True)
data.fillna(method='ffill', inplace=True)

plt.figure()
dendogram = sch.dendrogram(sch.linkage(data.loc[:, 'X1856':'X1976'], method='ward'))
plt.grid()
plt.show()
```

```python
# task 4
animal_data = pd.read_csv('datasets/animals.csv')
animal_data.rename(columns={'Unnamed: 0': 'City'}, inplace=True)
animal_data.fillna(method='backfill', inplace=True)

plt.figure()
animal_dendogram = sch.dendrogram(sch.linkage(animal_data.loc[:, 'war':'hai'], method='ward'))
plt.grid()
plt.show()
```
