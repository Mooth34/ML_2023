```python
import random
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC, SVR
from sklearn.metrics import mean_squared_error

from degree_dependence import degree_error_dependence, gamma_error_dependence
from display import display
```

# task 1

```python
train_data1 = pd.read_csv('datasets/svmdata1.txt', sep='\t')
test_data1 = pd.read_csv('datasets/svmdata1test.txt', sep='\t')
y_train1 = train_data1['Color']
y_train1.replace({'red': 0, 'green': 1}, inplace=True)
X_train1 = train_data1.drop(['Color'], axis=1).values  # train_data.loc[:, ['X1', 'X2']]

y_test1 = test_data1['Color']
y_test1.replace({'red': 0, 'green': 1}, inplace=True)
X_test1 = test_data1.drop(['Color'], axis=1).values  # test_data.loc[:, ['X1', 'X2']]
```

```python
clf = SVC(kernel='linear')
clf.fit(X_train1, y_train1)

print('Точноть для обучающей выборки: {}'.format(clf.score(X_train1, y_train1)))
print('Точность для тестовой выборки: {}'.format(clf.score(X_test1, y_test1)))
print('Опорные вектора: {}'.format(clf.support_vectors_))
```

```python
display(X_test1, y_test1, clf, 'SVM1 test', 'X1', 'X2')
display(X_train1, y_train1, clf, 'SVM1 train', 'X1', 'X2')
```

# task 6

```python
train_data6 = pd.read_csv('datasets/svmdata6.txt', sep='\t')
y = train_data6['Y'].values
X = train_data6['X'].values.reshape(-1, 1)
```

```python
# Максимальное значение epsilon
max_epsilon = 50
# Число испытаний для каждого epsilon
test_count = 50
# Массив значений epsilon
epsilon_values = [1 * (1 - float(i) / max_epsilon) for i in range(max_epsilon)]
# Зависимость ошибки от degree
mse_dependence = pd.Series(index=epsilon_values, dtype=float)
for epsilon in epsilon_values:
    mean_mse = 0.
    for _ in range(test_count):
        X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=random.randint(0, 1000))
        regr = SVR(kernel='rbf', C=1, epsilon=epsilon)
        regr.fit(X_train, y_train)
        y_pred = regr.predict(X_test)
        mean_mse += mean_squared_error(y_test, y_pred) / test_count
    mse_dependence[epsilon] = mean_mse
```

```python
plt.plot(mse_dependence, label='mse', marker='.', markersize=10)
plt.xlabel('Epsilon')
plt.ylabel('MSE')
plt.grid()
plt.show()
```
