# Задание 2. Генерация точек с нормальным распределением

```python
import random

import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score, GridSearchCV
from sklearn.naive_bayes import GaussianNB
```

## Генерация данных

```python
x1_1 = np.random.normal(10, 4, size=500)
x1_2 = np.random.normal(20, 3, size=500)
x2_1 = np.random.normal(14, 4, size=500)
x2_2 = np.random.normal(18, 3, size=500)
X = np.rot90(np.vstack((np.append(x2_1, x2_2), np.append(x1_1, x1_2))), k=-1)
y = np.append(np.zeros(500) - 1, np.ones(500))
```

# Построение классификатора

```python
gnb = GaussianNB()
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
score = cross_val_score(X=X, y=y, estimator=gnb, cv=kf, n_jobs=-1)
print('Scores: ' + str(score))
print('Mean score: ' + str(score.mean()))
```

 ## Графическое отображение результата

```python
plt.scatter(x1_1, x2_1, s=15, c='g')
plt.scatter(x1_2, x2_2, s=15, c='r')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid(True)
plt.show()
```
