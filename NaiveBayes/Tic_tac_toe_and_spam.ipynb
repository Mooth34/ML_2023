# Задание 1. Крестики-нолики и спам электронной почты

```python
import pandas as pd
import matplotlib.pyplot as plt
from accuracy_dependence import get_accuracy_dependence_on_the_sample_size, get_accuracy_dependence_on_the_test_size, normalize_data
```

## Чтение данных из файла

```python
data = pd.read_csv('datasets/Tic_tac_toe.txt', header=None)
y = data[9]
X = data.loc[:, '0':'8'].isin([1.0, 'x'])
X = normalize_data(X)
X
```

 ## Графическое отображение результата

```python
plt.plot(get_accuracy_dependence_on_the_test_size(X, y, int(y.index.size*0.6)), label='Tic_tac_toe',
marker='.', markersize=10)
plt.xlabel('Test size')
plt.ylabel('Accuracy')
plt.grid()
plt.show()
```

## Чтение данных из файла

```python
data = pd.read_csv('datasets/spambase.data', header=None)
y = data[57]
X = data.loc[:, '0':'56']
X = normalize_data(X)
X
```
